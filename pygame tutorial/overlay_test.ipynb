```python
import tkinter as tk
from tkinter import Menu
import pygame
import sys
from pygame import surfarray
import numpy as np

# Initialize Pygame
pygame.init()

# Create the Tkinter root window
root = tk.Tk()
root.title("Tkinter-Pygame Overlay Application")
root.geometry("640x480")  # Same size as the Pygame screen

# Create a canvas in Tkinter to hold the Pygame surface
canvas = tk.Canvas(root, width=640, height=480)
canvas.pack()

# Function to embed the Pygame display into the Tkinter window using a canvas
def update_canvas():
    """Update the Tkinter canvas with the Pygame surface"""
    pygame_surface = pygame.display.get_surface()
    
    # Convert the Pygame surface to a format Tkinter can use
    pygame_pixels = pygame.surfarray.array3d(pygame_surface)
    pygame_pixels = np.flip(pygame_pixels, axis=0)  # Flip the image to match the orientation
    photo = tk.PhotoImage(width=640, height=480)
    photo.putdata(list(pygame_pixels.reshape(-1, 3).flatten()))
    
    canvas.create_image(0, 0, anchor=tk.NW, image=photo)
    canvas.image = photo  # Store a reference to avoid garbage collection

# Create a Pygame screen surface
screen = pygame.Surface((640, 480))

# Function to run the Pygame loop
def run_pygame():
    clock = pygame.time.Clock()

    # Pygame loop
    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()

        screen.fill((255, 255, 255))  # Fill the screen with white
        pygame.draw.circle(screen, (255, 0, 0), (320, 240), 50)  # Draw a red circle

        # Update Tkinter canvas with the Pygame screen surface
        update_canvas()

        pygame.display.flip()  # Update the Pygame window (in case it's needed)
        clock.tick(60)  # Maintain 60 frames per second

# Create a menu bar for the Tkinter window
menubar = Menu(root)
root.config(menu=menubar)

# Add "File" menu with a "Cascade" option (currently empty)
file_menu = Menu(menubar, tearoff=0)
menubar.add_cascade(label="File", menu=file_menu)
file_menu.add_cascade(label="Cascade")  # Empty cascade menu

# Define the function to close the Tkinter window and quit Pygame
def close_app():
    pygame.quit()
    root.quit()

# Add a "Quit" option to the File menu
file_menu.add_separator()
file_menu.add_command(label="Quit", command=close_app)

# Start the Pygame loop in a separate thread
import threading
threading.Thread(target=run_pygame, daemon=True).start()

# Start the Tkinter main loop
root.mainloop()
```
