```python
import pygame
import random

# Initialize Pygame
pygame.init()

# Game settings
screen_width = 600
screen_height = 400
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("Dodge the Falling Blocks")

# Colors
WHITE = (255, 255, 255)
BLUE = (0, 0, 255)
RED = (255, 0, 0)

# Player settings
player_width = 50
player_height = 50
player_x = screen_width // 2 - player_width // 2
player_y = screen_height - player_height - 10
player_speed = 5

# Obstacle settings
obstacle_width = 50
obstacle_height = 50
obstacle_speed = 5
obstacles = []

# Game loop
running = True
clock = pygame.time.Clock()

while running:
    # Fill the screen with white
    screen.fill(WHITE)
    
    # Handle events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT and player_x > 0:
                player_x -= player_speed
            if event.key == pygame.K_RIGHT and player_x < screen_width - player_width:
                player_x += player_speed

    # Create new obstacles at random positions
    if random.random() < 0.02:
        new_obstacle_x = random.randint(0, screen_width - obstacle_width)
        obstacles.append([new_obstacle_x, -obstacle_height])
    
    # Move obstacles down
    for obstacle in obstacles:
        obstacle[1] += obstacle_speed
        if obstacle[1] > screen_height:
            obstacles.remove(obstacle)

    # Check for collisions
    for obstacle in obstacles:
        if (obstacle[1] + obstacle_height > player_y and 
            obstacle[1] < player_y + player_height and
            obstacle[0] + obstacle_width > player_x and
            obstacle[0] < player_x + player_width):
            running = False  # End the game if collision happens
    
    # Draw player
    pygame.draw.rect(screen, BLUE, (player_x, player_y, player_width, player_height))

    # Draw obstacles
    for obstacle in obstacles:
        pygame.draw.rect(screen, RED, (obstacle[0], obstacle[1], obstacle_width, obstacle_height))

    # Update the display
    pygame.display.update()

    # Set the frame rate
    clock.tick(60)

# Game over message
print("Game Over!")
pygame.quit()
```

```python
import tkinter as tk
from tkinter import ttk
import pygame
import random
import os
import sys

root = tk.Tk()
embed = ttk.LabelFrame(root, text="game", width=1280, height=720)
embed.grid(row=0, column=2)
root.update() # is this necessary?

os.environ['SDL_WINDOWID'] = str(embed.winfo_id())
os.environ['SDL_VIDEODRIVER'] = 'windib'
pygame.display.init()
screen = pygame.display.set_mode((640,480))
pygame.display.set_caption("Dodge the Falling Blocks")

pygame.init()

def close_app():
    pygame.quit()
    root.destroy()
    sys.exit()

WHITE = (255, 255, 255); BLUE = (0, 0, 255); RED = (255, 0, 0) # colors

# Player settings
player_width = 50
player_height = 50
player_x = screen_width // 2 - player_width // 2
player_y = screen_height - player_height - 10
player_speed = 5

# Obstacle settings
obstacle_width = 50
obstacle_height = 50
obstacle_speed = 5
obstacles = []

# Game loop
running = True
clock = pygame.time.Clock()

while running:
    # Fill the screen with white
    screen.fill(WHITE)
    
    # Handle events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    
    # Check key state
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and player_x > 0:
        player_x -= player_speed
    if keys[pygame.K_RIGHT] and player_x < screen_width - player_width:
        player_x += player_speed


    # Create new obstacles at random positions
    if random.random() < 0.02:
        new_obstacle_x = random.randint(0, screen_width - obstacle_width)
        obstacles.append([new_obstacle_x, -obstacle_height])
    
    # Move obstacles down
    for obstacle in obstacles:
        obstacle[1] += obstacle_speed
        if obstacle[1] > screen_height:
            obstacles.remove(obstacle)

    # Check for collisions
    for obstacle in obstacles:
        if (obstacle[1] + obstacle_height > player_y and 
            obstacle[1] < player_y + player_height and
            obstacle[0] + obstacle_width > player_x and
            obstacle[0] < player_x + player_width):
            running = False  # End the game if collision happens
    
    # Draw player
    pygame.draw.rect(screen, BLUE, (player_x, player_y, player_width, player_height))

    # Draw obstacles
    for obstacle in obstacles:
        pygame.draw.rect(screen, RED, (obstacle[0], obstacle[1], obstacle_width, obstacle_height))

    # Update the display
    pygame.display.update()
    root.update()

    # Set the frame rate
    clock.tick(60)

    # Keep Tkinter GUI responsive
    try:
        root.update()
    except TclError:
        # This happens if the Tk window is closed directly
        running = False

close_app()

# Game over message
print("Game Over!")
```
