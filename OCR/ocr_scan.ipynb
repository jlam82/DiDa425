## __1__ The Functions

```python
def PIL_to_cv2(img_obj):
    """
    Helper function.

    Converts a PIL image object into a cv2 image object.
    """
    cv2_img = np.array(img_obj) # convert the PIL image object into a numpy array
    cv2_img = cv2_img[:, :, ::-1] # convert RGB to BGR (opencv format)
    
    return cv2_img
```

```python
def grayscale(image):
    """
    Helper function.

    Grayscales the image to assist in OCR.
    """
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
def OCR(pdf_path):
    """
    Perform Optical Character Recognition of a PDF file by
    first converting to images. pdf2image converts the pdf
    pages to PIL objects (nice).
    """
    imgs = convert_from_path(pdf_path)

    for n, i in enumerate(imgs):
        print(f"""
==========
PAGE {n+1}
==========
        """)
        cv2_img = PIL_to_cv2(i)
        img_gray = grayscale(cv2_img)

        print(pytesseract.image_to_string(img_gray))
        print("") # for formatting
```

```python
OCR("to_scan//Nintendo Annual Data.pdf")
```
